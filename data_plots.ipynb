```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate data

We generate uniform $x$ in $[0, 1]$ and $y$ is sampled from $N(f(x), \sigma)$, where $f(x) = \exp(\sin(2\pi x))$ and $\sigma = 0.1$.

```python
f = lambda x: np.exp(np.sin(2*np.pi*x))

np.random.seed(0)

N = 100
x = np.random.rand(N, 1)
y = np.random.randn(N, 1) * 0.1 + f(x)

with open('data.txt', 'w') as f:
    for i in range(N):
        f.write(f'{x[i, 0]} {y[i, 0]}\n')
```

```python
y_apx = np.zeros(N)
with open('run.txt') as f:
    y_apx = np.array([float(line) for line in f])


plt.plot(x, y, 'o', label='data')
plt.plot(x, y_apx, 'rs', label='approximation')
```
